```python
from langchain_community.document_loaders import PyPDFDirectoryLoader
from pathlib import Path

def load_documents():
    pdf_directory = Path("C:/Users/carlf/Documents/GitHub/LawIntelAfrica/data/01_raw")
    document_loader = PyPDFDirectoryLoader(str(pdf_directory))
    return document_loader.load()
doc = load_documents()
doc
```

```python
import pandas as pd

def document_to_dict(doc):
    """
    Convertit un objet Document en dictionnaire pour le DataFrame.
    """
    doc_dict = {
        "text": doc.page_content
    }
    
    # Ajouter toutes les métadonnées disponibles
    for key, value in doc.metadata.items():
        doc_dict[key] = value
    
    return doc_dict

# Appliquer la fonction à tous les documents
data = [document_to_dict(docc) for docc in doc]

# Créer le DataFrame
df = pd.DataFrame(data)

# Afficher les premières lignes du DataFrame
print(df.head())
```

```python
df['document_name'] = df['source'].apply(
        lambda x: x.split('01_raw\\')[1].split('.')[0]
    )
```
